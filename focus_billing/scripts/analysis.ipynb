```python
import pandas as pd

df = pd.read_csv("focus_azure_billing/focus_billing.csv")
```

```python
display(df)
```

```python
df.info()
```

```python
total_billed_cost = df['BilledCost'].sum()

print(f"Total Billed Cost: ${total_billed_cost:.2f}")
```

```python
df['BillingMonth'] = df['BillingPeriodStart'].dt.to_period('M')
monthly_trend = df.groupby('BillingMonth')['BilledCost'].sum().reset_index()

display(monthly_trend)
```

```python
df['BillingWeek'] = df['BillingPeriodStart'].dt.to_period('W')  # Extract week
weekly_trend = df.groupby('BillingWeek')['BilledCost'].sum().reset_index()
weekly_trend = weekly_trend.sort_values(by='BillingWeek')

display(weekly_trend)
```

```python
service_cost = df.groupby('ServiceCategory')['BilledCost'].sum().reset_index()
service_cost = service_cost.sort_values(by='BilledCost', ascending=False)

display(service_cost)
```

```python
resource_cost = df.groupby('ResourceType')['BilledCost'].sum().reset_index()
resource_cost = resource_cost.sort_values(by='BilledCost', ascending=False)

display(resource_cost)
```

```python
reservation_analysis = df.groupby('ChargeCategory')['BilledCost'].sum().reset_index()
reservation_analysis = reservation_analysis.sort_values(by='BilledCost', ascending=False)

display(reservation_analysis)
```
